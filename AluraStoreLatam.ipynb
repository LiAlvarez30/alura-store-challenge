# **Importación de datos**



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

# **Exploración de las 4 tiendas**


```python
import pandas as pd

tiendas = {
    "Tienda 1": tienda,
    "Tienda 2": tienda2,
    "Tienda 3": tienda3,
    "Tienda 4": tienda4
}

for nombre, df in tiendas.items():
    print("\n========================")
    print(nombre)
    print("========================")
    print("Shape (filas, columnas):", df.shape)
    print("\nColumnas:")
    print(list(df.columns))
    print("\nTipos de datos (dtypes):")
    print(df.dtypes)
    print("\nValores nulos (top 10):")
    display(df.isna().sum().sort_values(ascending=False).head(10))
```

# **Unir las 4 tiendas en un solo DataFrame**

```python
# Creamos copias y agregamos una columna "Tienda" para identificar el origen
t1 = tienda.copy()
t1["Tienda"] = "Tienda 1"

t2 = tienda2.copy()
t2["Tienda"] = "Tienda 2"

t3 = tienda3.copy()
t3["Tienda"] = "Tienda 3"

t4 = tienda4.copy()
t4["Tienda"] = "Tienda 4"

# Unimos todo en un solo DataFrame
df_all = pd.concat([t1, t2, t3, t4], ignore_index=True)

# Convertimos Fecha de Compra a fecha real (datetime)
df_all["Fecha de Compra"] = pd.to_datetime(df_all["Fecha de Compra"], dayfirst=True, errors="coerce")

# Chequeos rápidos
print("Shape df_all:", df_all.shape)
display(df_all["Tienda"].value_counts())
print("Fechas inválidas (NaT):", df_all["Fecha de Compra"].isna().sum())

df_all.head()
```

# **1. Facturación total por tienda**



**1.1 Crear la tabla de facturación**



```python
# Facturación total por tienda (suma de precios)
facturacion = df_all.groupby("Tienda")["Precio"].sum().sort_values(ascending=False)

print("Facturación total por tienda:")
facturacion.apply(lambda x: f"{x:,.0f}")
```

**Gráfico (barras)**

```python
import matplotlib.pyplot as plt

# Aseguramos el orden: de mayor a menor
facturacion = df_all.groupby("Tienda")["Precio"].sum().sort_values(ascending=False)

# Convertimos a "millones" para que se lea mejor
fact_millones = facturacion / 1_000_000

fig, ax = plt.subplots(figsize=(10, 5))

# Colores automáticos (se ven bien y no son todos iguales)
colors = plt.get_cmap("tab10").colors
bars = ax.barh(fact_millones.index, fact_millones.values, color=colors[:len(fact_millones)])

ax.invert_yaxis()  # mayor arriba
ax.set_title("Facturación total por tienda", fontsize=14, pad=12)
ax.set_xlabel("Facturación (millones)", fontsize=11)
ax.set_ylabel("")

ax.grid(axis="x", alpha=0.25)

# Etiquetas al final de cada barra (2 decimales)
for b in bars:
    val = b.get_width()
    ax.text(val + 1, b.get_y() + b.get_height()/2, f"{val:,.2f} M",
            va="center", fontsize=10)

plt.tight_layout()
plt.show()
```

**Insight (Facturación):**  
La **Tienda 1** presenta la mayor facturación total, mientras que la **Tienda 4** registra la menor.  
Esto sugiere que, desde la perspectiva de ingresos, la Tienda 4 podría ser candidata a venta si el resto de métricas no compensa esta diferencia.


# **2. Ventas por categoría**

**2.1 Top categorías por tienda (por cantidad de ventas)**

```python
# Conteo de ventas por categoría y tienda
ventas_categoria = (
    df_all.groupby(["Tienda", "Categoría del Producto"])
    .size()
    .reset_index(name="Cantidad")
)

# Top 5 categorías por tienda (desempate consistente por nombre)
top5_cat = (
    ventas_categoria.sort_values(["Tienda", "Cantidad", "Categoría del Producto"],
                                 ascending=[True, False, True])
    .groupby("Tienda")
    .head(5)
)

top5_cat
```

**Gráfico (barras horizontales por tienda)**

```python
import matplotlib.pyplot as plt

tiendas_orden = ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"]
cmap = plt.get_cmap("tab10")

fig, axes = plt.subplots(2, 2, figsize=(14, 9))
axes = axes.flatten()

for ax, t in zip(axes, tiendas_orden):
    data = top5_cat[top5_cat["Tienda"] == t].sort_values("Cantidad", ascending=True)

    # Color por categoría (dentro de cada tienda)
    colors = [cmap(i) for i in range(len(data))]

    bars = ax.barh(data["Categoría del Producto"], data["Cantidad"], color=colors)
    ax.set_title(f"Top 5 categorías más vendidas — {t}", fontsize=12)
    ax.set_xlabel("Cantidad de ventas")
    ax.set_ylabel("")
    ax.grid(axis="x", alpha=0.25)

    # Etiquetas de cantidad al final de cada barra
    for b in bars:
        ax.text(b.get_width() + 1, b.get_y() + b.get_height()/2, f"{int(b.get_width())}",
                va="center", fontsize=10)

    # Leyenda (cuadro) con colores
    handles = [plt.Line2D([0], [0], marker='s', linestyle='', markersize=10, color=colors[i])
               for i in range(len(data))]
    labels = list(data["Categoría del Producto"])
    ax.legend(handles, labels, title="Categoría (color)", loc="center left", bbox_to_anchor=(1.02, 0.5), frameon=True)

plt.tight_layout()
plt.show()

```

**Insight (Categorías):**  
En las cuatro tiendas, las categorías con mayor volumen de ventas se concentran principalmente en **Muebles** y **Electrónicos**, seguidas por **Juguetes** y **Electrodomésticos**.  
Esto indica un patrón de demanda similar entre tiendas; por lo tanto, para diferenciar rendimiento conviene complementar con facturación, calificaciones y costos de envío.


# **3. Calificación promedio de la tienda**

# **4. Productos más y menos vendidos**

**4.1 Top 5 productos más vendidos por tienda**

```python
import pandas as pd
import matplotlib.pyplot as plt

# Conteo de ventas por producto y tienda
ventas_prod = (
    df_all.groupby(["Tienda", "Producto"])
    .size()
    .reset_index(name="Cantidad")
)

top5_por_tienda = (
    ventas_prod
    .sort_values(["Tienda", "Cantidad", "Producto"], ascending=[True, False, True])  # orden estable
    .groupby("Tienda")
    .head(5)
)

print("Top 5 productos más vendidos por tienda:")

top5_por_tienda


```

**4.3 Top por tienda + más/menos por tienda**

```python
tiendas_orden = ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"]

# Top 3 por tienda (desempate estable por nombre)
top3 = (
    ventas_prod.sort_values(["Tienda", "Cantidad", "Producto"], ascending=[True, False, True])
    .groupby("Tienda")
    .head(3)
    .copy()
)

# Más vendido por tienda
mas_vendido = (
    ventas_prod.sort_values(["Tienda", "Cantidad", "Producto"], ascending=[True, False, True])
    .groupby("Tienda")
    .head(1)
    .rename(columns={"Producto":"Producto_mas", "Cantidad":"Cant_mas"})
)

# Menos vendido por tienda
menos_vendido = (
    ventas_prod.sort_values(["Tienda", "Cantidad", "Producto"], ascending=[True, True, True])
    .groupby("Tienda")
    .head(1)
    .rename(columns={"Producto":"Producto_menos", "Cantidad":"Cant_menos"})
)

extremos = mas_vendido.merge(menos_vendido, on="Tienda")
extremos = extremos.set_index("Tienda").loc[tiendas_orden].reset_index()

display(extremos)
```

**Gráfico por tienda + más/menos vendido**

```python
fig, axes = plt.subplots(2, 2, figsize=(14, 8))
axes = axes.flatten()

color_mas = "#d62728"    # rojo
color_menos = "#ffbf00"  # amarillo

for ax, t in zip(axes, tiendas_orden):
    row = extremos[extremos["Tienda"] == t].iloc[0]

    productos = [row["Producto_mas"], row["Producto_menos"]]
    cantidades = [row["Cant_mas"], row["Cant_menos"]]
    colores = [color_mas, color_menos]

    bars = ax.bar(productos, cantidades, color=colores)

    ax.set_title(f"{t} — Más vs Menos vendido", fontsize=12)
    ax.set_ylabel("Cantidad de ventas")
    ax.grid(axis="y", alpha=0.25)
    ax.tick_params(axis="x", rotation=20)

    # Etiquetas arriba de cada barra
    for b in bars:
        ax.text(b.get_x() + b.get_width()/2, b.get_height() + 0.8,
                f"{int(b.get_height())}", ha="center", va="bottom", fontsize=10)

    # Leyenda simple
    ax.legend(["Más vendido", "Menos vendido"], frameon=True)

plt.tight_layout()
plt.show()
```

**Gráfico productos mas vendidos por tienda**

```python
cmap = plt.get_cmap("tab10")

fig, axes = plt.subplots(2, 2, figsize=(14, 9))
axes = axes.flatten()

for ax, t in zip(axes, tiendas_orden):
    data = top3[top3["Tienda"] == t].sort_values("Cantidad", ascending=True)

    colors = [cmap(i) for i in range(len(data))]
    bars = ax.barh(data["Producto"], data["Cantidad"], color=colors)

    ax.set_title(f"Top 3 productos más vendidos — {t}", fontsize=12)
    ax.set_xlabel("Cantidad de ventas")
    ax.set_ylabel("")
    ax.grid(axis="x", alpha=0.25)

    # Etiquetas al final de cada barra
    for b in bars:
        ax.text(b.get_width() + 0.8, b.get_y() + b.get_height()/2,
                f"{int(b.get_width())}", va="center", fontsize=10)

    # Cuadro (leyenda) con producto y color
    handles = [plt.Line2D([0], [0], marker='s', linestyle='', markersize=10, color=colors[i])
               for i in range(len(data))]
    labels = list(data["Producto"])
    ax.legend(handles, labels, title="Producto (color)", loc="center left",
              bbox_to_anchor=(1.02, 0.5), frameon=True)

plt.tight_layout()
plt.show()
```

## Insight — Productos más y menos vendidos

Los gráficos muestran qué productos impulsan el volumen de ventas en cada tienda (Top) y cuáles tienen menor salida (Bottom).  
Esto permite identificar **productos estrella** (alta demanda) y **productos de bajo desempeño** (posibles candidatos a rotación o promoción).  
La comparación “más vendido vs menos vendido” también ayuda a ver si la tienda depende de pocos productos o si tiene una demanda más distribuida.

# **6. Tabla resumen + “score” para decidir la tienda a vender**

```python

import pandas as pd

# Recalculamos por seguridad (si ya existen, igual está ok)
facturacion = df_all.groupby("Tienda")["Precio"].sum()
prom_calif = df_all.groupby("Tienda")["Calificación"].mean()
prom_envio = df_all.groupby("Tienda")["Costo de envío"].mean()

resumen = pd.DataFrame({
    "Facturación total": facturacion,
    "Calificación promedio": prom_calif,
    "Envío promedio": prom_envio
})

# Formato: facturación en millones, 2 decimales en calificación y envío
resumen_fmt = resumen.copy()
resumen_fmt["Facturación (M)"] = (resumen_fmt["Facturación total"] / 1_000_000).round(2)
resumen_fmt["Calificación promedio"] = resumen_fmt["Calificación promedio"].round(2)
resumen_fmt["Envío promedio"] = resumen_fmt["Envío promedio"].round(2)

# Nos quedamos con columnas presentables
resumen_fmt = resumen_fmt[["Facturación (M)", "Calificación promedio", "Envío promedio"]].sort_values("Facturación (M)", ascending=False)

resumen_fmt

```

**6.1 Ranking y score final**

```python
resumen_score = resumen.copy()

# Rankings: 1 = mejor
resumen_score["Rank Facturación"] = resumen_score["Facturación total"].rank(ascending=False)
resumen_score["Rank Calificación"] = resumen_score["Calificación promedio"].rank(ascending=False)
resumen_score["Rank Envío"] = resumen_score["Envío promedio"].rank(ascending=True)

# Score total: más alto = peor (suma de ranks)
resumen_score["Score final (más alto = peor)"] = (
    resumen_score["Rank Facturación"] +
    resumen_score["Rank Calificación"] +
    resumen_score["Rank Envío"]
)

resumen_score = resumen_score.sort_values("Score final (más alto = peor)", ascending=False)

# Formato presentable (2 decimales donde corresponde)
resumen_score_out = resumen_score.copy()
resumen_score_out["Facturación (M)"] = (resumen_score_out["Facturación total"] / 1_000_000).round(2)
resumen_score_out["Calificación promedio"] = resumen_score_out["Calificación promedio"].round(2)
resumen_score_out["Envío promedio"] = resumen_score_out["Envío promedio"].round(2)

resumen_score_out = resumen_score_out[[
    "Facturación (M)",
    "Calificación promedio",
    "Envío promedio",
    "Rank Facturación",
    "Rank Calificación",
    "Rank Envío",
    "Score final (más alto = peor)"
]]

resumen_score_out
```
